# Manipulating Images as ndarrays

```python
import matplotlib.pyplot as plt
from PIL import Image
```

```python
sea_beach = Image.open("beach.JPG")
sea_beach = np.array(sea_beach)
print(f'Data Type: {type(sea_beach)}')
plt.imshow(sea_beach)
```

```python
#display the upside down image
plt.imshow(np.flip(sea_beach))
```

```python
#rotate 90
plt.imshow(np.rot90(sea_beach))
```

```python
solar = 255 - sea_beach
plt.imshow(solar)
```

# Cobverting the image to black and white

```python
#The values in my image range from 0 to 255.
sea_beach
```

```python
#Need to convert RGB to sRGB valuues in range 0 to 1
sRGB = sea_beach/255
sRGB
```

#### Grayscale Liner Equation:  
$Y_{linear}  = 0.2126R_{linear} +0.7152G_{linear} +0.0722B_{linear}$

```python
grey_vals = np.array([0.2126, 0.7152,0.0722])
grey_vals
```

```python
grey_img = sRGB @ grey_vals
plt.imshow(grey_img, cmap='gray')
```
